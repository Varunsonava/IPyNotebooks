```python
import numpy as np
```

```python
training_data=np.loadtxt('training_boston_x_y_train.csv',delimiter=',')
test_x=np.loadtxt('test_boston_x_test.csv',delimiter=',')
```

```python
training_data.shape
```

```python
X=training_data[:,0:13]
X.shape
Y=training_data[:,13]
Y.shape
```

```python
X=np.c_[X,np.ones(X.shape[0])]
```

```python
test_x=np.c_[test_x,np.ones(test_x.shape[0])]
```

```python
test_x.shape
```

```python
# Feature Scaling on training data
from sklearn.preprocessing import StandardScaler,MinMaxScaler

from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(2)
X=poly.fit_transform(X)
test_x=poly.transform(test_x)

mscaler=MinMaxScaler()
X=mscaler.fit_transform(X)
test_x=mscaler.transform(test_x)

scaler=StandardScaler()
X=scaler.fit_transform(X)
test_x=scaler.transform(test_x)
```

```python
def cost(slope):
    total_cost=0
    M=X.shape[0]
    for i in range(X.shape[0]):
        x=X[i]
        y=Y[i]
        total_cost+=(1/M)*((y-(slope*x).sum())**2)
    return total_cost
        
    
def sg(lr,slope):
    M=len(X)
    new_slope=np.zeros(X.shape[1])
    for i in range(X.shape[0]):
        output=Y[i]
        for j in range(X.shape[1]):
            curr=X[i][j]
            new_slope[j]=(-2/M)*(output-(slope*X[i]).sum())*curr
        slope=slope-lr*new_slope
    return slope
        
def gd():
    lr=0.001
    numIter=50000
    slope=np.zeros(X.shape[1])
    for i in range(numIter):
        slope=sg(lr,slope)
        print(i," Cost: ",cost(slope))
    print(slope)
    return slope
```
