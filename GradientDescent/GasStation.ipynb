### FEATURE SCALING

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler

mscaler=MinMaxScaler()
X=mscaler.fit_transform(X)
test_x=mscaler.transform(test_x)

stdscaler=StandardScaler()
X=stdscaler.fit_transform(X)
test_x=stdscaler.transform(test_x)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(2)
X=poly.fit_transform(X)
test_x=poly.transform(test_x)
```

```python
def cost(slope):
    total_cost=0
    M=X.shape[0]
    for i in range(X.shape[0]):
        x=X[i]
        y=Y[i]
        total_cost+=(1/M)*((y-(slope*x).sum())**2)
    return total_cost
        
    
def sg(lr,slope):
    M=len(X)
    new_slope=np.zeros(X.shape[1])
    for i in range(X.shape[0]):
        output=Y[i]
        for j in range(X.shape[1]):
            curr=X[i][j]
            new_slope[j]=(-2/M)*(output-(slope*X[i]).sum())*curr
        slope=slope-lr*new_slope
    return slope
        
def gd():
    lr=0.001
    numIter=10000
    slope=np.zeros(X.shape[1])
    for i in range(numIter):
        slope=sg(lr,slope)
        print(i," Cost: ",cost(slope))
    print(slope)
    return slope
```
