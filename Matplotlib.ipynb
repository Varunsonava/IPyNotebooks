```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line and scatter plots

```python
x=[1,2,3]
y=[4,5,6]
plt.scatter(x,y)
```

```python
plt.plot(x,y,'b') # for the line, first character decides the line colour
plt.scatter(x,y) # for the points
plt.show()
```

```python
# for 2 lines
x2=[6,7,8]
y2=[9,10,11]
plt.scatter(x2,y2)
plt.plot(x2,y2)
plt.scatter(x,y)
plt.plot(x,y)
```

## Different Views of line

```python
# Dashed line
plt.plot(x,y,'b--') # for the line, first character decides the line colour
plt.scatter(x,y) # for the points
plt.show()
```

```python
# Plus point line
plt.plot(x,y,'b+') # for the line, first character decides the line colour
plt.scatter(x,y) # for the points
plt.show()
```

```python
# For carrot points
plt.plot(x,y,'b^') # for the line, first character decides the line colour
plt.scatter(x,y) # for the points
plt.show()
```

```python
# For pentagon shaped points
plt.plot(x,y,'bp') # for the line, first character decides the line colour
plt.scatter(x,y) # for the points
plt.show()
```

```python
# Dash dot representation
plt.plot(x,y,'b-.') # for the line, first character decides the line colour
plt.scatter(x,y) # for the points
plt.show()
```

```python
# Plot x^3
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y,"bo")
```

## Customizing Graph

```python
# Plot x^3
x=np.arange(0,5,0.2)
y=x**3
y2=x**2
plt.plot(x,y2,color='red',marker='o',linewidth=2,label='x^2')
plt.plot(x,y,color='blue',marker='o',linewidth=2,label='x^3') # marker are the shape of the points
plt.ylabel('Y')
plt.xlabel('X')
plt.title('Matplotlib Demo')
plt.legend() # to show the labels of the plot
plt.axis([0,10,0,120]) # for specifying the x and y axis range
plt.grid() # this will create a grid for us
plt.text(8,20,"Text",fontsize=20) # to show the text at random points on graph
```

## Pie Graph

```python
# Pie graphs
labels=["A","B","C","D"]
sizes=[2,3,6,4]
colors=["blue","purple","red","green"]
explode=[0.1,0.2,0.1,0]
plt.title('Pie Graph')
plt.axis("equal")
plt.pie(sizes,labels=labels,colors=colors,autopct="%.2f%%",explode=explode,counterclock=False,startangle=100) # counterclock is by default anti-clockwise
plt.show() # only shows the plot and nothing else
```
